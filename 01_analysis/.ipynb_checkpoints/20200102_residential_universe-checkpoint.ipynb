```python
# making subset frame from clean spatil joined deeds_C
deeds_Cs = deeds_C

# selecting just post 2000 deeds
deeds_Cs['sort_year'] = pd.to_numeric(deeds_Cs['sort_year'])
deeds_Cs = deeds_Cs.loc[deeds_Cs['sort_year']>1999]

# selecting investors by grantee name
C_inv = deeds_Cs['grantees'].str.contains('LLC|LP|LIMITED PARTNERSHIP|CORP|CORPORATION|INC|COMPANY|LIMITED PARTNE|L P|L L C|PARTNERSHIP|LTD|TRUST',regex=True)

# converting investors series to dataframe
C_inv = C_inv.to_frame()

# merging with deeds
deeds_C_inv = deeds_Cs.merge(C_inv, left_index=True, right_index = True)
deeds_C_inv = deeds_C_inv.dropna(subset=['grantees_x'])
```

```python
# creating a count column to allow grouping and count 
# for investor and owner by year 
deeds_C_inv['inv'] = deeds_C_inv['grantees_y'].apply(lambda x: 1 if x == True else 0)
deeds_C_inv['occ'] = deeds_C_inv['grantees_y'].apply(lambda x: 1 if x == False else 0)
```

```python
# making a test set of investors to run the nearest neighbor calculation on 

testInv = deeds_C_inv[deeds_C_inv['grantees_x'].str.contains('CRESPO JOHN; CRESPO LAURA|DG MANAGEMENT')==True]
```

```python
columnsTouse = ['grantees_x','sort_year','objectid']
testInvColumns = testInv[columnsTouse]
testInvColumns['x']=0
testInvColumns['y']=testInvColumns['sort_year']
testInvColumns= testInvColumns.drop('sort_year',1)
```

```python
testInvColumns.head()
```

```python
testInvColumns.drop('objectid',1).groupby('grantees_x').count()
```

```python
# nearest neighbor calculation

def nn(x): 
    nbrs = NearestNeighbors(n_neighbors=2, algorithm='auto', metric='euclidean').fit(x)
    distances, indices = nbrs.kneighbors(x)
    return distances, indices

nns = testInvColumns.drop('objectid',1).groupby('grantees_x').apply(lambda x: nn(x.as_matrix()))

groups = testInvColumns.groupby('grantees_x')
nn_rows = []
names = testInvColumns['grantees_x'].unique()
for i, nn_set in enumerate(nns):
    name = names[i]
    group = groups.get_group(name)
    for j, tup in enumerate(zip(nn_set[0], nn_set[1])):
        nn_rows.append({'grantees_x': name,
                        'objectid': group.iloc[j]['objectid'],
                        'nearest_neighbour': group.iloc[tup[1][1]]['objectid'],
                        'euclidean_distance': tup[0][1]})
nn_df = pd.DataFrame(nn_rows).set_index('grantees_x')

# leaves us with a dataframe that shows the distance to the 'nearest' deed for each deed (grouped by grantee)
# need next to identify the miniumum value for this euclidian distance. if it is less than 2 then that means that 
# there are two deeds within three years of each other
```
