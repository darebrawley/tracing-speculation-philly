### 1A. subsetting just residential categories

looking at just the subcategories of the residential use types

```python
# checking the level 3 categories in each type of residential use
Res_med = landuse.loc[landuse['C_DIG2DESC']=='Residential Medium']
Res_low = landuse.loc[landuse['C_DIG2DESC']=='Residential Low']
Res_hi = landuse.loc[landuse['C_DIG2DESC']=='Residential High']
Res_mixed = landuse.loc[landuse['C_DIG2DESC']=='Commercial Mixed Residential']
med_list = Res_med['C_DIG3DESC'].unique()
low_list = Res_low['C_DIG3DESC'].unique()
hi_list = Res_hi['C_DIG3DESC'].unique()
mixed_list = Res_mixed['C_DIG3DESC'].unique()
```

```python
# # cleaning data to remove erronious '235' descripton which corresponds to "residential rowhouse"
# res_B = res_B.replace({'C_DIG3DESC':'235'},"Residential Rowhouse")
```

```python
# creating a more narrow set of residential parcels
# just low and medium residential uses
res_C = landuse.loc[(landuse['C_DIG2DESC']=='Residential Medium') | 
                    (landuse['C_DIG2DESC']=='Residential Low')]
```

```python
# print(np.subtract(res_A.shape,res_C.shape))
# print(np.subtract(res_B.shape, res_C.shape))
```

```python
# dropping pacels with na values for their geometry
# res_A = res_A.dropna(subset=['geometry'])
# res_B = res_B.dropna(subset=['geometry'])
res_C = res_C.dropna(subset=['geometry'])
```

```python
# # reprojecting the residential parcels
# res_A = res_A_na.to_crs({'init':'epsg:2272'})
# res_B = res_B_na.to_crs({'init':'epsg:2272'})
```

```python
# reprojecting residential low/med only category
res_C = res_C.to_crs({'init':'epsg:2272'})
```
