```python
high_inv = msa.loc[msa['% of Home Purchases by Investors']>0.10]
```

```python
annual_inv = high_inv.drop(columns=['% of Low-Priced Homes Purchased by Investors',
       '% of Mid-Priced Homes Purchased by Investors',
       '% of High-Priced Homes Purchased by Investors',
       '% Point Change 5-Year', '% Point Change 10-Year',
       '% Point Change 20-Year', '% Point Change 5-Year Low',
       '% Point Change 10-Year Low', '% Point Change 20-Year Low'])
```

```python
annual_inv.head()
```

```python
annual_inv.plot(kind='line',x='year',y='% of Home Purchases by Investors')
```
