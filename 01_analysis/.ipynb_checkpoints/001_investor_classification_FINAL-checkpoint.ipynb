```python
import geopandas as gpd
import pandas as pd
pd.set_option('display.max_columns', 80)
from shapely.geometry import Point, LineString
from geopandas import GeoDataFrame
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.neighbors import NearestNeighbors
import datetime
```

## 1. Selecting Deeds that fall within residential lots

```python
# load residential universe C
res_C = gpd.read_file("../00_data/PROCESSED/land_use/res_C.geojson")
```

```python
# load real estate tax transfer data
rtt = pd.read_csv("../00_data/RAW/philadelphia/realestate_tax_transfers/RTT_SUMMARY.csv")
```

```python
# selecting just deeds
deeds = rtt.loc[rtt['document_type'] == "DEED"]

# adding a year column from the document date
deeds["sort_year"] = deeds['display_date'].astype(str).str[:4]
```

```python
deeds.shape
```

```python
# points from xy
deeds_points = [Point(xy) for xy in zip(deeds.lng, deeds.lat)]
deedsgeo = GeoDataFrame(deeds, geometry=deeds_points)
```

```python
# selecting only deeds with non zero state taxes
deedsgeo = deedsgeo.loc[deedsgeo["state_tax_amount"] > 0]

# and that also have an address field
deedsgeo= deedsgeo.loc[deedsgeo["address_low"]>0]
```

```python
# setting the projection for the lat lon coordinates 
deedsgeo.crs = {'init' :'epsg:4326'}
```

```python
# reprojecting to: EPSG:2272: NAD83 / Pennsylvania South (ftUS)
deedsgeo = deedsgeo.to_crs({'init':'epsg:2272'})
```

```python
deeds_C = gpd.sjoin(deedsgeo,res_C, how="inner",op="intersects")
```

```python
deeds_C.shape
```

```python
deeds_C.plot(figsize=(20,20),markersize=.5)
```

## 2. Identifying Deeds with likely investors as the grantee based on grantee name
A. first tag all investor named entities  
    (LLC|LP|LIMITED PARTNERSHIP|CORP|CORPORATION|INC|COMPANY|LIMITED PARTNE|L P|L L C|PARTNERSHIP|LTD|TRUST)
B. then find all investors who purchased more than 2 properties in any three year span  


```python
# making subset frame from clean spatial joined deeds_C
deeds_Cs = deeds_C

# selecting just post 2000 deeds
deeds_Cs['sort_year'] = pd.to_numeric(deeds_Cs['sort_year'])
deeds_Cs = deeds_Cs.loc[deeds_Cs['sort_year']>1999]

# selecting investors by grantee name
C_inv = deeds_Cs['grantees'].str.contains(
    'LLC|LP|LIMITED PARTNERSHIP|CORP|CORPORATION|INC|COMPANY|LIMITED PARTNE|L P|L L C|PARTNERSHIP|LTD|TRUST'
    ,regex=True)

# converting investors series to dataframe
C_inv = C_inv.to_frame()

# merging with deeds
deeds_C_inv = deeds_Cs.merge(
    C_inv, left_index=True, right_index = True)
deeds_C_inv = deeds_C_inv.dropna(subset=['grantees_x'])
```

```python
# creating a count column to allow grouping and count 
# for investor and owner by year 
deeds_C_inv['inv'] = deeds_C_inv['grantees_y'].apply(lambda x: 1 if x == True else 0)
deeds_C_inv['occ'] = deeds_C_inv['grantees_y'].apply(lambda x: 1 if x == False else 0)
```

```python
# dropping duplicate object ids from the dataset
deeds_C_inv = deeds_C_inv.drop_duplicates(subset="objectid")
```

```python
deeds_C_inv['inv']==1
```

## 3. Classifying investor purchases by identifying grantees who purchased more than 2 properties in any one year period

### 3.a. Identifying multiple purchasers from the full set of non-investor classified deeds:

```python
# finding multiple purchasers from the full set of non-investor classified deeds
# selecting from just the non llc llp deeds
occ = deeds_C_inv.loc[deeds_C_inv['occ']==1]
```

```python
# getting all deeds from purchasers who are associated with more than one deed
multiple_purchasers = occ.groupby(
    "grantees_x").filter(lambda s: s['objectid'].count()>=2)

# converting date to datetime format
multiple_purchasers['display_date'] = multiple_purchasers[
    'display_date'].astype('datetime64[ns]')

# dropping unneeded columns
columns_multi = ['objectid','grantors','grantees_x',
                 'display_date','sort_year','street_address']
multiple_purchase_simple = multiple_purchasers[columns_multi]

# dropping duplicate objectids so the groupby doesnt' throw an error
# multiple_purchase_simple_nodup = multiple_purchase_simple.drop_duplicates(subset='objectid')

multiple_purchase_simple[
    'elapsed'] = multiple_purchase_simple.sort_values(
    ['grantees_x','display_date']).groupby(
    'grantees_x')['display_date'].diff()

# converting timedelta column to an int type 
# to make it possible to filter below
multiple_purchase_simple[
    'elapsed_days'] = multiple_purchase_simple[
    'elapsed'].dt.days

```

```python
# filtering for deeds from grantees who purchased two properties within 1 year of each other
multiple_investors = multiple_purchase_simple.groupby('grantees_x').filter(
    lambda x: x['elapsed_days'].min() <= 365)
multiple_investors['multi_invest'] = 1
```

```python
multiple_investors
```

## 4. merging multiple purchase investors back with full set of deeds

compiling a comprehensive classification of investor purchases 

```python
deeds_C_multi = deeds_C_inv.merge(
    multiple_investors, left_on="objectid", right_on="objectid",how="left")
```

```python
deeds_C_multi.head()
```

```python
deeds_C_multi["investor_all"] = deeds_C_multi[
    "inv"] + deeds_C_multi["multi_invest"].fillna(0)
deeds_C_multi['owner_occup'] = deeds_C_multi[
    'investor_all'].apply(lambda x: 0 if x == 1 else 1)
```

```python
# converting datatypes for save

# d_typ = {'display_date_y': str, 'elapsed': int}
deeds_for_save = deeds_C_multi
```

```python
deeds_for_save['elapsed'] = deeds_for_save['elapsed'].astype(int)
deeds_for_save['display_date_y'] = deeds_for_save['display_date_y'].astype(str)
```

```python
# saving final list of grantees
with open("../00_data/PROCESSED/realestate_tax_transfers/rtt_investors_classified.geojson",
          'w') as f:
    f.write(deeds_for_save.to_json())
```
