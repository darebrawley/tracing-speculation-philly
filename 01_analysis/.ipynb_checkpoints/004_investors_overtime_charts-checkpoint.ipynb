```python
# %matplotlib inline
import geopandas as gpd
import pandas as pd
pd.set_option('display.max_columns', 100)
from geopandas import GeoDataFrame
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import mapclassify as mc
# import pysal
```

```python
# deeds = gpd.read_file("../00_data/PROCESSED/realestate_tax_transfers/rtt_investors_classified.geojson")
# deeds.crs = {'init':'epsg:2272'}

deeds = gpd.read_file("../00_data/PROCESSED/realestate_tax_transfers/rtt_investors_classified_2021.geojson")
deeds.crs = {'init':'epsg:2272'}
```

```python
deeds['count_field'] = 1
deeds['year'] = deeds['sort_year_x'].astype(str)
deeds.head()
```

```python
# grouping to be able to plot

total = deeds.groupby('year').agg({'count_field':'sum','inv':'sum','investor_all':'sum','occ':'sum'})
total['inv_nonLLC'] = total['investor_all']-total['inv']
total['inv_per'] = (total['investor_all']/total['count_field'])
total['inv_per_nonLLC'] = (total['inv_nonLLC']/total['count_field'])
total['inv_per_LLC'] = (total['inv']/total['count_field'])
```

```python
total
```

```python
f, (ax1) = plt.subplots(1,1,figsize=(15, 10))
# sns.set(style="white" )
sns.lineplot(data=total['inv','investor_all','occ'], ax=ax1)
```

## comparing density of likely investor purchases



```python
grouped_ct_10_18 = gpd.read_file("../00_data/PROCESSED/grouped_ct_deeds10_18.geojson")
grouped_ct_00_10 = gpd.read_file("../00_data/PROCESSED/grouped_ct_deeds00_10.geojson")
```

```python
grouped_ct_10_18 = grouped_ct_10_18.drop(['GEOID10','geometry','res_parcel_num'],axis=1)
```

```python
grouped_ct_00_10 = grouped_ct_00_10.drop(['STFID','geometry','res_parcel_num'],axis=1)
```

```python
mean1018= grouped_ct_10_18.mean()
```

```python
sd1018 = grouped_ct_10_18.std()
```

```python
mean1018 = mean1018.append(grouped_ct_00_10.mean())
sd1018 = sd1018.append(grouped_ct_00_10.std())
```

```python
stats = pd.concat([mean1018,sd1018],axis=1)
names = ['mean','sd']
stats.columns = names
stats = stats.reset_index()
stats['year'] = stats['index'].str[-4:]
stats['cat'] = stats['index'].str[:-4]
```

```python
stats_shaped = stats.pivot(index='year',columns='cat',values=['mean','sd'])
```

```python
stats_shaped.columns = ['_'.join((col[1], col[0])) for col in stats_shaped.columns]
```

```python
stats_line = stats_shaped.reset_index()
```

```python
stats_line = stats_line.astype({'year':'int64'})
```
